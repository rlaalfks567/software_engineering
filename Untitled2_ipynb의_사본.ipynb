```
#-*-coding:utf-8-*-
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('./data/sample02.jpg');

# mask생성
mask = np.zeros(img.shape[:2],np.uint8)
mask[100:300,100:400] = 255

# 이미지에 mask가 적용된 결과
masked_img = cv2.bitwise_and(img,img,mask=mask)

# 원본 이미지의 히스토그램
hist_full = cv2.calcHist([img],[1],None,[256],[0,256])

# mask를 적용한 히스트로그램
hist_mask = cv2.calcHist([img],[1],mask,[256],[0,256])

plt.subplot(221),plt.imshow(img,'gray'),plt.title('Origianl Image')
plt.subplot(222),plt.imshow(mask,'gray'),plt.title('Mask')
plt.subplot(223),plt.imshow(masked_img,'gray'),plt.title('Masked Image')

# red는 원본이미지 히스토그램, blue는 mask적용된 히스토그램
plt.subplot(224),plt.title('Histogram')
plt.plot(hist_full,color='r'),plt.plot(hist_mask,color='b')
plt.xlim([0,256])

plt.show()
```

```
# OS modules
# 도구 호출 / 폴더 생성 삭제를 위해서
import os
import shutil

#import Pythn Image Library (PIL)
# Image modules
from PIL import ImageFont
from PIL import Image
from PIL import ImageDraw

# create/delete our temp files folder

if os.path.exists('frames'):   #frames 라는 폴더가 존재하면?
    shutil.rmtree('frames')    #폴더 삭제
os.mkdir('frames')  #현재 위치에 새 폴더 'frames' 생성

# load in the background image
#Imgae.open 이미지 열기 / Image.new(모드, 사이즈, 컬러) 새로운 이미지파일 생성
img_background = Image.open('./data/sample05.jpg')
img = Image.new("RGBA", img_background.size, (0, 0, 0, 255))
# PIL.Image.new(mode, size, color=0)

# set up colours and vars
x = 10
y = 1
silver = (100, 100, 100, 255)
purple = (100, 0, 200, 255)
random = (200, 150, 175, 0)
text = '''Hi?!''' # 이미지에 띄우고싶은 텍스트를 설정

# load the font
font = ImageFont.truetype("./data/THE복숭아틴트.ttf", 500)  # 폰트 불러오기
draw = ImageDraw.Draw(img)

# add text to each frame
for N in range(0, 24):   # 24프레임 이미지 생성을 위해 range(0, 24) 사용
    x += 2*N
    y += 4*N  # 텍스트를 y축으로 이동시키기위해서
    random = (10+10*N, 150, 175, 0)
    img.paste(img_background, (0, 0)) # img에 (0, 0) 좌표에 img_background를 복사
    draw.text((x+4, y+4), text, purple, font=font)
    draw.text((x+2, y+2), text, silver, font=font)
    draw.text((x, y), text, random, font=font)
    # 생성된 이미지 저장
    img.save("./frames/{}.png".format(str(N).zfill(3)))

# output the result
# AVI 파일 생성 및 GIF 변환
os.system('ffmpeg -framerate 24 -i frames/%03d.png -c:v ffv1 -r 24 -y myTextAvi.avi')
os.system('ffmpeg -y -i myTextAvi.avi myTextGif.gif')

# clean up
shutil.rmtree('frames') # 작업폴더 frames 삭제

plt.imshow(img)
plt.show()
```
